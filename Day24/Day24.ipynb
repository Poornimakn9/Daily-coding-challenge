```python
rides = (
    (101, "Ravi", "Bangalore", 250, 4.5),
    (102, "Anjali", "Hyderabad", 180, 4.2),
    (103, "Kiran", "Bangalore", 300, 4.8),
    (104, "Meena", "Chennai", 220, 4.0),
    (105, "Arjun", "Bangalore", 150, 3.9)
)
```

```python
##Task1
##Count total rides
##Find how many ride records are present.
total_rides = len(rides)
print("Total rides:", total_rides)
```

```python
#Task2
##Extract all rides from Bangalore
##Print only tuples where city = "Bangalore"

bangalore_rides = []

for ride in rides:
    if ride[2] == "Bangalore":
        bangalore_rides.append(ride)

print(bangalore_rides)
```

```python
#Task3
##Find total revenue from Bangalore
##Sum of fares where city is Bangalore

total_revenue = sum(ride[3] for ride in rides if ride[2] == "Bangalore")

print("Total Bangalore Revenue:", total_revenue)
```

```python
#Task4
##Find highest rated driver
##Print the tuple with maximum rating

highest_rated = max(rides, key=lambda x: x[4])

print("Highest Rated Driver:", highest_rated)
```

```python
#Task5
##Average rating per city
city_ratings = {}

for ride in rides:
    city = ride[2]
    rating = ride[4]

    if city not in city_ratings:
        city_ratings[city] = []

    city_ratings[city].append(rating)

# Calculate averages
avg_rating_per_city = {
    city: round(sum(ratings)/len(ratings), 1)
    for city, ratings in city_ratings.items()
}

print(avg_rating_per_city)
```
